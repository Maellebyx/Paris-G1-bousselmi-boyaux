```python
 
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
 
#Get the data on temperatures and first check
df = pd.read_csv("/content/CLIMAT RISK EXCEL.csv")  
df.dtypes
df.head()
df.info()
df.isnull().sum()
df.columns
df.describe()
 
"""The df.head function is used to check that the data in the table is output"""
 
#reorganise the data
df = df.rename(columns={'Country':'country'})
df = df.rename(columns={'cri_score':'score'})
df.head(33)
 
"""We used the choropleth package for its ease of use. The map allowed us to visualise each country in the world with relevant colours and according to their climate risk index. """
 
# create map
fig = px.choropleth(data_frame=df,
                    locations="country",
                    locationmode="country names",
                    color="score",
                    scope="world")
fig.show()
 
"""with this part we could have the map on a html file so that it is easier to visualize and manipulate"""
 
# reorganise the data
fig.update_layout(title_text="Climate risks and economic losses",
                  coloraxis_colorbar_title="score")
fig.write_html("europe.html",auto_open=True)
 
"""We first tried to frame our data but this did not sort it in ascending order. Indeed the expression selects the rows in the df DataFrame where the value of the "score" column is either greater than 12.17 or less than 26, but this leads to a selection of all rows if all "score" values are either greater than 12.17 or less than 26.
 
To be able to select rows where the value of the "score" column is both greater than 12.17 and less than 26, we have seen that we need to use the & operator instead of the | operator
"""
 
#create a chart with the ten less risky countries
df_top=df[(df.loc[:,"score"]>12.170000)&(df.loc[:,"score"]<26.000000)]
df_top = df_top.sort_values(by='score', ascending=True)
 
"""To visualise our data we found it relevant to use the "seaborn" package and add a relevant title. """
 
sns.barplot(x="score",y="country",data=df_top,palette="crest")
plt.title("10 less risky countries regarding the climate risk index")
plt.show()


```
